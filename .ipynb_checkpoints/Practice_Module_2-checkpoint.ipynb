```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('companies.csv')
```

```python
#Q_1 prep
df['state'].value_counts()
```

```python
#Q_1 Creating order df by states most to least
df_states_counts = df['state'].value_counts().index
```

```python
#Q_1 create a countplot that shows a count of how many of the companies in the data set are from each state
plt.figure(figsize=(16, 8))
sns.countplot(x='state', data=df, order=df_states_counts)
plt.show()
```

```python
#Q_3 prep
df.groupby('Industry')['job_openings'].sum().sort_values(ascending=False)
```

```python
#Q_3 assigning new data frame
df_industry = df.groupby('Industry')['job_openings'].sum().sort_values(ascending=False).reset_index()
```

```python
#Q_2 create a bar chart that shows a sum of how many total job openings there are in each industry.
plt.figure(figsize=(22, 8))
sns.barplot(x='Industry', y='job_openings', data=df_industry)
plt.show()
```

```python
#Q_5 "line chart that shows the average growth rate for each founded year"
df.groupby('founded')['growth_percentage'].mean().round(2)
```

```python
#Q_5 new data frame
df_growthrate_year = df.groupby('founded')['growth_percentage'].mean().round(2).reset_index()
```

```python
#Q_5 struggled with output, had to label to get line chart
sns.lineplot(x='founded', y='growth_percentage', data=df_growthrate_year, marker='o')
plt.xticks(ticks=range(1910, 2026, 10))
plt.title('Average Growth Percentage by Founding Year')
plt.xlabel('Founded Year')
plt.ylabel('Growth Percentage')
plt.show()
```

```python
#Q_7 "create a scatter plot that compares the number of employees to estimated revenues for each company"
sns.scatterplot(x='employees', y='estimated_revenues', data=df)
plt.xlabel('Employees')
plt.ylabel('Estimated Revenue (in Billions)')
plt.show()
```

```python
#Q_9 "create a histogram that measures the number of companies founded in each year"
sns.histplot(x='founded', data=df)
plt.show()
```
